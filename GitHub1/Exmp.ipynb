```python
import pandas as pd
import numpy as np
dados_cidades = pd.DataFrame({
    "Nome" : ["Ana", "Richard", "Francisco", "Roberto", "Valdirene"],
    "Cidade" : ["São Paulo", np.nan, "Curitiba", np.nan, "Belém"]
})

dados_cidades["Cidade_Preenchida"] = dados_cidades["Cidade"].fillna("Não informado")
dados_cidades
```

```python
dados_temperaturas = pd.DataFrame({
    "Dia" : ["Segunda", "Terça", "Quarta", "Quinta", "Sexta"],
    "Temperatura" : [30, np.nan, np.nan, 28, 27]
})

dados_temperaturas["Preenchido_bfill"] = dados_temperaturas["Temperatura"].bfill()
dados_temperaturas
```
